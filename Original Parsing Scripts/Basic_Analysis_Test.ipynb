```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np
import math
from struct import *
import base64
import matplotlib.pyplot as plt
from csv import reader
import glob
```

```python
f2 = 'test_data/091322_163220_david_test_VIDGAME1_RESUME.csv'
f1 = 'test_data/091322_162451_david_test_VIDGAME1_CONTD.csv'
f0 = 'test_data/091322_160117_david_test_VIDGAME1.csv'
```

```python
#so this is 2 mini-sessions in one session; f0 and f1/f2, while playing guitar

def get_data(file):
    logs, blinks, acc, gyro, thermal, user, watch = [], [], [], [], [], [], []
    with open(file, 'r') as read_obj:
        csv_reader = reader(read_obj)
        for row in csv_reader:
            if   (row[0]=='l'):
                logs.append(row[1:])

            elif (row[0]=='g'):
                if   (row[2]=='b'):
                    blinks.append(row[1:])
                elif (row[2]=='t'):
                    thermal.append(row[1:])
                elif (row[2]=='a'):
                    acc.append(row[1:])
                elif (row[2]=='g'):
                    gyro.append(row[1:])
            
            elif (row[0]=='u'):
                user.append(row[1:])
                
            elif (row[0]=='w'):
                watch.append(row[1:])
                
    return logs, blinks, acc, gyro, thermal, user, watch

s0 = get_data(f0)

s1 = get_data(f1)
s2 = get_data(f2)
#combine s2 into s1 since it was a pause
[s1[i].extend(s2[i]) for i in range(len(s1))]
del s2

#now we have s0 and s1
```
